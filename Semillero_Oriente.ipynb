Operaciones Logicas

```python
games_me = {"GTA V", "Spiderman Miles Morales", "CSGO"}
games_others = {"GTA V", "Halo Infinite", "CSGO"}

print("Mis juegos:", games_me)
print("Juegos de otros:", games_others)

en_comun = games_me & games_others
print("\n Juegos que ambos tenemos:", en_comun)

nuevos_para_mi = games_others - games_me
print("\n Juegos que otros tienen y yo no:", nuevos_para_mi)

games_me.add("Red Dead Redemption 2")
print("\n Ahora mis juegos son:", games_me)

todos_juegos = games_me | games_others
print("\nTodos los juegos juntos:", todos_juegos)
```

# Start HW

```python
heart_disease = fetch_ucirepo(id=45)
df = pd.concat([heart_disease.data.features, heart_disease.data.targets], axis=1)

print("Total de pacientes:", len(df))
print("\nAs칤 se ve el dataset:")
print(df.head())
```

### Limpieza de Datos

```python
print("\nCu치ntos datos vac칤os hay:", df.isnull().sum().sum())

# Elimino las filas con datos vac칤os porque son pocas
# y es mejor trabajar con datos completos

df = df.dropna()

print("Pacientes despu칠s de limpiar:", len(df))

# Cambio el nombre de la columna 'num' para que sea m치s claro
df = df.rename(columns={'num': 'enfermedad'})

# Creo una columna que dice si tiene enfermedad (1) o no (0)
df['tiene_enfermedad'] = (df['enfermedad'] > 0).astype(int)
```

### Filtros

```python
print("\nFiltros")

# Filtro 1: solo mayores de 50
mayores_50 = df[df['age'] > 50]
print("Mayores de 50:", len(mayores_50))

# Filtro 2: separar por sexo
mujeres = df[df['sex'] == 0]
hombres = df[df['sex'] == 1]
print("Mujeres:", len(mujeres))
print("Hombres:", len(hombres))

# Filtro 3: colesterol alto
alto_colesterol = df[df['chol'] > 240]
print("Con colesterol alto:", len(alto_colesterol))
```

### Promedos y Sumas

```python
print("\nPromedios")
print("Edad promedio:", round(df['age'].mean(), 1))
print("Colesterol promedio:", round(df['chol'].mean(), 1))
print("Presi칩n promedio:", round(df['trestbps'].mean(), 1))
print("Porcentaje con enfermedad:", round(df['tiene_enfermedad'].mean() * 100, 1), "%")
```

### Agrupaciones

```python
print("\nPor Sexo")
por_sexo = df.groupby('sex')['tiene_enfermedad'].mean() * 100
print(por_sexo)

# Creo grupos de edad
df['grupo_edad'] = pd.cut(df['age'], bins=[0, 50, 60, 100], labels=['Menos de 50', '50 a 60', 'M치s de 60'])

print("\nPor Edad")
por_edad = df.groupby('grupo_edad')['tiene_enfermedad'].mean() * 100
print(por_edad)
```

### Tabla Summary

```python
print("\nTabla Summary")
resumen = df.groupby(['grupo_edad', 'sex']).agg({
    'age': 'count',
    'chol': 'mean',
    'tiene_enfermedad': 'mean'
})
print(resumen)
```

# Graficas

### Grafica 1 Edades

```python
plt.hist(df['age'], bins=15, color='lightblue', edgecolor='black')
plt.title('Edades de los pacientes')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.show()

print("\nGrafica 1: La mayor칤a tiene entre 50 y 60 a침os")
```

### Grafica 2 Colesterol

```python
sin_enfermedad = df[df['tiene_enfermedad'] == 0]['chol']
con_enfermedad = df[df['tiene_enfermedad'] == 1]['chol']

plt.hist([sin_enfermedad, con_enfermedad], bins=15,
         label=['Sin enfermedad', 'Con enfermedad'],
         color=['lightgreen', 'lightcoral'])
plt.title('Colesterol')
plt.xlabel('Colesterol')
plt.ylabel('Cantidad')
plt.legend()
plt.show()

print("\nGrafica 2: El colesterol es parecido en ambos grupos")
print("Sin enfermedad promedio:", round(sin_enfermedad.mean(), 1))
print("Con enfermedad promedio:", round(con_enfermedad.mean(), 1))
```

### Grafica 3 Hombres y Mujeres

```python
conteo = df.groupby(['sex', 'tiene_enfermedad']).size().unstack()
conteo.plot(kind='bar', color=['lightgreen', 'lightcoral'])
plt.title('Enfermedad por sexo')
plt.xlabel('Sexo (0=Mujer, 1=Hombre)')
plt.ylabel('Cantidad')
plt.legend(['Sin enfermedad', 'Con enfermedad'])
plt.xticks(rotation=0)
plt.show()

print("\nGrafica 3: Los hombres tienen m치s casos de enfermedad")
```

### Grafica 4 Por Edad

```python
porcentaje_edad = df.groupby('grupo_edad')['tiene_enfermedad'].mean() * 100
porcentaje_edad.plot(kind='bar', color='orange')
plt.title('Enfermedad por edad')
plt.xlabel('Edad')
plt.ylabel('Porcentaje (%)')
plt.xticks(rotation=0)
plt.show()

print("\nGrafica 4: A mayor edad, m치s enfermedad")
```
